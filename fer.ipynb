```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('happyboy.jpg')
if img is None:
    print("Error: Image not loaded properly!")
    exit()

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load Haar cascade
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Detect faces
faces = faceCascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Draw rectangles around detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 3)

# Convert color for matplotlib display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show the result using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Hide axes
plt.show()
```

```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

# Ensure 'img' is properly loaded before using it
predictions = DeepFace.analyze(img)

# Define font style for text
font = cv2.FONT_HERSHEY_SIMPLEX  

# Add text to the image with RED color
cv2.putText(
    img,
    predictions[0]['dominant_emotion'],  # Get the dominant emotion
    (50, 50),  # Position of the text
    font,  # Font type
    1,  # Font scale
    (255, 0, 0),  # Font color (Red)
    2,  # Thickness
    cv2.LINE_AA  # Line type
)

# Convert BGR (OpenCV format) to RGB (Matplotlib format)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image inside Jupyter Notebook
plt.imshow(img_rgb)
plt.axis("off")  # Hide axis
plt.show()
```

```python
import cv2
import threading
from deepface import DeepFace

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    raise IOError("Cannot open webcam")

emotion_result = [{"dominant_emotion": "Loading..."}]

def analyze_emotion(frame):
    global emotion_result
    try:
        emotion_result = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)
    except:
        emotion_result = [{"dominant_emotion": "No Face"}]

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.resize(frame, (480, 360))  # Smaller frame for faster processing
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = faceCascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=7, minSize=(50, 50))

    if frame_count % 10 == 0:
        thread = threading.Thread(target=analyze_emotion, args=(frame,))
        thread.start()

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    cv2.putText(frame, emotion_result[0]['dominant_emotion'], (50, 50), cv2.FONT_HERSHEY_SIMPLEX,
                1, (0, 0, 255), 2, cv2.LINE_AA)

    cv2.imshow("Face Emotion Recognition", frame)
    
    frame_count += 1

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
